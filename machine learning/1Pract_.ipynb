```python
x=2 #starting pt
lr = 0.01
precision = 0.000001 #algo kdhi thambvaycha
previous_step_size = 1
max_iter = 10000
iters = 0 
gf = lambda x:(x+3)**2
```

```python
while precision < previous_step_size and iters<max_iter:
    prev = x
    x=x-lr*gf(prev)
    previous_step_size = abs(x-prev)
    iters+=1
    print('iter:',iters,'value:',x)
```
