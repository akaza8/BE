```python
import cv2
import matplotlib.pyplot as plt
```

# Read image dataset

```python
img = cv2.imread("MyPhoto.jpg")
```

```python
plt.imshow(img, cmap='gray') 
plt.title('Original Image')
plt.show()
```

# Sobel edge detection

```python

image_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
image_blur = cv2.GaussianBlur(image_gray, (3,3), 0)
sobelxy = cv2.Sobel(src=image_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)
sobelxy = cv2.convertScaleAbs(sobelxy)

plt.imshow(sobelxy, cmap='gray') 
plt.title('Sobel Edge')
plt.show()
```

# Canny edge detection

```python

image_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
image_blur = cv2.GaussianBlur(image_gray, (3,3), 0)
edges = cv2.Canny(image_blur, 50, 50)

plt.imshow(edges, cmap='gray') 
plt.title('Canny Edge')
plt.show()
```

# Laplacian edge detection

```python
image_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
image_blur = cv2.GaussianBlur(image_gray, (3,3), 0)
edges = cv2.Laplacian(image_blur, cv2.CV_64F)
edges = cv2.convertScaleAbs(edges)

plt.imshow(edges, cmap='gray') 
plt.title('Laplacian Edge')
plt.show()
```
