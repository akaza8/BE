# Modular code

```python
def sample_image(image, size, interpolation=None):
    if interpolation is not None:
        subsampled_image = cv2.resize(image, size, interpolation)
    else:
        subsampled_image = cv2.resize(image, size)
    return subsampled_image
```

```python
# subsampling
sizes = [1024, 512, 256, 128, 64, 32]

for index, size in enumerate(sizes):
    img = cv2.imread("MyPhoto.jpg")
    subsampled_img = sample_image(img, (size, size))
    plt.subplot(1, 6,index+1)
    plt.title(str(size) + "*" + str(size))
    plt.axis('off')
    plt.imshow(subsampled_img)
```

```python
# resampling
sizes = [32, 64, 128, 256, 512, 1024]

for index, size in enumerate(sizes):
    if size == 32:
        sampled_img = sample_image(img, (32, 32))
        plt.subplot(1, 6,index+1)
        plt.title(str(size) + "*" + str(size))
        plt.axis('off')
        plt.imshow(sampled_img)
        continue
    plt.subplot(1, 6,index+1)
    plt.title(str(size) + "*" + str(size))
    plt.axis('off')
    plt.imshow(sampled_img)
    sampled_img = sample_image(sampled_img, (size, size))
```
